## Analyse des ensembles de niveau 

#### Question 1:

Soit $c \in \mathbb{R}$.
On suppose que la fonction $f:\mathbb{R}^2 \to \mathbb{R}$ est continue et vérifie
$f(x_1, x_2) \to +\infty$ quand $\|(x_1,x_2)\| \to +\infty$.
Que peut-on dire de l'ensemble de niveau $c$ de $f$ ?

#### Réponse:

On appelle *ensemble de niveau* $c \in \mathbb{R}$ d'une fonction $f : \mathbb{R}^2 \to \mathbb{R}$  l'ensemble du plan
$$
\{(x_1,x_2) \in \mathbb{R}^2 \; | \; f(x_1, x_2) = c\}.
$$

On définit $E$ comme l'ensemble de niveau de $c$ de $f$. 
- Cet ensemble est l'image réciroque du singleton $\{c\}$, un fermé de $\mathbb{R}$, par une fonction continue. On peut donc affirmer que E est un fermé.
- Par ailleurs, $E$ est borné. En effet, étant donné que $f$ vérifie $f(x_1, x_2) \to +\infty$ quand $\|(x_1,x_2)\| \to +\infty$, on peut toujours trouver un majorant : $$\forall A \in \mathbb{R}, \exists m \in \mathbb{R}, \|x\| \geq m \implies f(x) \geq A$$ 
- Enfin, il est possible d'affirmer que $E$ est non vide, à condition de prendre:
$$
c \geq inf\{f(x_1, x_2) | \; (x_1,x_2) \in \mathbb{R}^2\}
$$
En effet, comme $f$ est continue et que $f(x_1, x_2) \to +\infty$ quand $\|(x_1,x_2)\| \to +\infty$, on peut utiliser le théorème des valeurs intermédiaires ici. 


Dans la suite la fonction $f$ est supposée continûment différentiable. On suppose également que le gradient $\nabla f$ ne s'annule pas dans un voisinage du point $x_0 = (x_{10}, x_{20}) \in \mathbb{R}^2$. On pose alors
$$
p(x_1, x_2) := \frac{\partial_2 f(x_0)}{\|\nabla f(x_0)\|} (x_1 - x_{10}) -
\frac{\partial_1 f(x_0)}{\|\nabla f(x_0)\|} (x_2 - x_{20}).
$$

#### Question 2:
Comment interpréter géométriquement le terme $p(x_1,x_2)$ ?


#### Réponse:
On remarque que l'on peut écrire $p(x_1,x_2)$ comme le produit scalaire des deux vecteurs suivants:
$$
\frac{1}{\|\nabla f(x_0)\|}\begin{pmatrix}
   \partial_2 f(x_0) \\
   -\partial_1 f(x_0)
\end{pmatrix}
et 
 \begin{pmatrix}
   x_1-x_{10} \\
   x_2-x_{20}
\end{pmatrix}
$$
On peut alors interpréter cette expression de la façon suivante:
$p(x_1,x_2)$ est le déterminant des vecteurs:
$$
\overrightarrow{XX_0}=\begin{pmatrix}
   x_1-x_{10} \\
   x_2-x_{20}
\end{pmatrix}
et 
\frac{1}{\|\nabla f(x_0)\|}\begin{pmatrix}
   \partial_1 f(x_0) \\
   \partial_2 f(x_0)
\end{pmatrix}
$$
Celà correspond donc à l'aire signée du parallèlogramme engendré par ces vecteurs. Il est nul si ces deux vecteurs sont colinéaires. Ainsi, $p(x_1,x_2)$ est la projection du vecteur $\overrightarrow{XX_0}$ sur la tangente à la courbe de niveau $f(x)=c$ en 
$(x_{10},x_{20})$.

#### Question 3:

Montrer que dans un voisinage ouvert de $x_0$, on peut paramétriser l'ensemble de niveau $c:=f(x_0)$ au moyen de $p(x_1,x_2),$ c'est-à-dire qu'il existe un $\varepsilon > 0$ et une fonction (continûment différentiable) $\gamma :\left]-\varepsilon,\varepsilon \right[ \to \mathbb{R}^2$ tels que dans un voisinage ouvert de $x_0,$ $f(x_1,x_2) = c$ si et seulement si $(x_1, x_2) = \gamma(t)$ où $t = p(x_1, x_2)$.

#### Réponse:
On introduit la fonction $F$, telle que: $F : \mathbb{R}^2 \to \mathbb{R}^2$, qui se définit par:
$$
 F(x_1,x_2,t):= \begin{pmatrix}
   f(x_1,x_2)-c \\
   p(x_1,x_2)-t
\end{pmatrix}
$$
où $t=p(x_1,x_2)$. En notant, $\overrightarrow{x}=(x_1,x_2)$, on peut alors écrire:
$$
F(\overrightarrow{x},t):= \begin{pmatrix}
   f(\overrightarrow{x})-c \\
   p(\overrightarrow{x})-t
\end{pmatrix}
$$

On souhaite appliquer le théorème des fonctions implicites à une telle fonction, on vérifie donc les hypothèses nécessaires:
- $F$ doit être continuement différentiable. Or, par hypothèse, on a que $f$ est continuement différentiable donc $F_1$ est continuement différentiable par opérations. Par ailleurs, 
- $\partial_x F$ doit être inversible en $(\overrightarrow{x_0},t_0)$. On écrit:
$$
\partial_x F(\overrightarrow{x_0},t_0)=\begin{pmatrix}
   \partial_{x_1} (f(\overrightarrow{x_0})-c) &  \partial_{x_2} (f(\overrightarrow{x_0})-c)\\
   \partial_{x_1} (p(\overrightarrow{x_0})-t_0) & \partial_{x_2} (p(\overrightarrow{x_0})-t_0)
\end{pmatrix}
$$
On calcule le déterminant de la matrice précédente pour vérifier qu'elle est inversible:
$\Delta= \partial_{x_1} (f(\overrightarrow{x_0})-c) \times \partial_{x_2} (p(\overrightarrow{x_0})-t_0)-\partial_{x_1} (p(\overrightarrow{x_0})-t_0) \times \partial_{x_2} (f(\overrightarrow{x_0})-c)$
Or, par définition de $p(x_1,x_2)$:
$$
\partial_{x_2} (p(\overrightarrow{x_0})-t_0)=\partial_{x_2} p(\overrightarrow{x_0})=-\frac{\partial_{x_1} f(\overrightarrow{x_0})}{\|\nabla f(x_0)\|}
$$ 
et:
$$
\partial_{x_1} (p(\overrightarrow{x_0})-t_0)=\partial_{x_1} p(\overrightarrow{x_0})=\frac{\partial_{x_2} f(\overrightarrow{x_0})}{\|\nabla f(x_0)\|}
$$
d'où:
$$
\Delta=-\frac{\partial_{x_1} f(\overrightarrow{x_0})}{\|\nabla f(x_0)\|}^2-\frac{\partial_{x_2} f(\overrightarrow{x_0})}{\|\nabla f(x_0)\|}^2
$$
Or, d'après l'énoncé, le gradient $\nabla f$ ne s'annule pas dans un voisinage du point $x_0 = (x_{10}, x_{20}) \in \mathbb{R}^2$, et donc $_Delta$ non plus. Ainsi, la deuxième hypothèse du théorème des fonctions implicites est validée. 

Ainsi, comme le point $(\overrightarrow{x_0},t_0)$ vérifie $F(\overrightarrow{x_0},t_0)=0$, il existe des voisinages ouverts $U$ de $\overrightarrow{x_0}$ et $\left]-\varepsilon,\varepsilon \right[ $ de $t_0$ tels que $U \times \left]-\varepsilon,\varepsilon \right[  \subset \mathbb{R}^2$ et une unique fonction implicite $\gamma \to \mathbb{R}^2$, continuement différentiable, telle que pour tout t $\in \left]-\varepsilon,\varepsilon \right[$ et tout $x _in U$:
$$
\boxed{F(\overrightarrow{x},t)=0 \Leftrightarrow x=\gamma(t) \Leftrightarrow (x_1,x_2)=\gamma(t)}
$$

#### Question 4

Montrer que pour tout $t \in \left]-\varepsilon, \varepsilon \right[$ :

  - le vecteur $\gamma'(t)$ est non nul (il fournit donc une tangente au chemin $\gamma$),

  - est orthogonal à $\nabla f(\gamma(t))$.

#### Réponse:

- Le Théorème des fonctions implicites utilisé à la question précédente indique que la différentielle de $\gamma$ est donnée pout tout $t \in \left]-\varepsilon, \varepsilon \right[$ par:
$$
d\gamma(t)=-(\partial_x F(\overrightarrow{x},t))^{-1}.\partial_t F(\overrightarrow{x},t) \: où \: x=\gamma(t)
$$
Ainsi, on a:
$$
\gamma'(t)=\begin{pmatrix}
   \partial_{x_1} (f(\overrightarrow{x})-c) &  \partial_{x_2} (f(\overrightarrow{x})-c)\\
   \partial_{x_1} (p(\overrightarrow{x})-t) & \partial_{x_2} (p(\overrightarrow{x})-t)
\end{pmatrix}^{-1}.\partial_t F(\overrightarrow{x},t)
$$
Or,
$$
\partial_t F(\overrightarrow{x},t)=\begin{pmatrix}
   \ 0 \\
   -1
\end{pmatrix}
$$
et la matrice inverse d'une matrice $A=\begin{pmatrix}
   \ a & b \\
   c & d
   \end{pmatrix}$ de dimensions $2 \times 2$ inversible est:
$$
A^{-1}=\frac{1}{\Delta} \times \begin{pmatrix}
   \ d & -b \\
   -c & a
   \end{pmatrix}
$$
d'où:
$$
\gamma'(t)=-\frac{1}{\Delta} \times \begin{pmatrix}
   \partial_{x_2} (p(\overrightarrow{x})-t) & -\partial_{x_2} (f(\overrightarrow{x})-c)\\
   -\partial_{x_1} (p(\overrightarrow{x})-t) & \partial_{x_1} (f(\overrightarrow{x})-c)
\end{pmatrix} \times \begin{pmatrix}
   \ 0 \\
   -1
\end{pmatrix}=-\frac{1}{\Delta} \times \begin{pmatrix}
   -\partial_{x_2} f(\overrightarrow{x}) \\
   -\partial_{x_1} f(\overrightarrow{x})
\end{pmatrix}
$$
Or, d'après l'énoncé, le gradient $\nabla f$ ne s'annule pas dans un voisinage du point $x_0 = (x_{10}, x_{20}) \in \mathbb{R}^2$ donc $\begin{pmatrix}
   \partial_{x_2} f(\overrightarrow{x}) \\
   -\partial_{x_1} f(\overrightarrow{x})
\end{pmatrix}$ n'est pas nul et $\frac{1}{\Delta}$ ne s'annule pas non plus d'après la question précédente. On a donc bien montré que le vecteur $\gamma'(t)$ est non nul, il fournit donc une tangente au chemin $\gamma$ 

- Pour montrer que $\gamma'(t)$ est orthogonal à $\nabla f(\gamma(t))$, on calcule le produit scalaire de ces deux vecteurs:
$$
 \langle \gamma'(t),\nabla f(\gamma(t))\rangle = -\frac{1}{\Delta} \times \langle \begin{pmatrix}
   \partial_{x_2} f(\overrightarrow{x}) \\
   -\partial_{x_1} f(\overrightarrow{x})
\end{pmatrix},\begin{pmatrix}
   \partial_{x_1} f(\overrightarrow{x}) \\
   \partial_{x_2} f(\overrightarrow{x})
\end{pmatrix}\rangle=0
$$
On peut donc en conclure que ces deux vecteurs sont orthogonaux.

## Construction des courbes de niveau

### Méthode de Newton

Afin de trouver des points sur des courbes de niveau, on souhaite implémenter un algorithme de résolution des équations de la forme $F(x) = 0$ où $F: \mathbb{R}^2 \to \mathbb{R}^2$, par la méthode de Newton, amorcée par un point de référence $x_0 \in \mathbb{R}^2$.

L'implémentation prendra la forme générale donnée ci-dessous. L'appel $x, y = Newton(F, x0, y0, eps, N)$ devra renvoyer le couple des coordonnées d'un point $(x, y)$ qui est une solution approchée de $F(x,y)=0$ à proximité de $(x_0, y_0)$. La qualité de l'approximation produite est déterminée par le réel positif $\varepsilon > 0$, une approximation de la distance entre le point $(x,y)$ renvoyé par la fonction et la solution exacte. L'algorithme itératif s'interrompra si aucune approximation de la précision souhaitée n'a été trouvée en $N$ étapes.

#### Tâche 1:

Implémenter la fonction `Newton` en complétant le canevas ci-dessous

```python
# Third-Party Libraries
# ---------------------

# Autograd & Numpy
import autograd
import autograd.numpy as np

# Pandas
#import pandas as pd

# Matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 12] # [width, height] (inches). 

# Jupyter & IPython
from IPython.display import display

#on implémente les outils proposés en début d'énoncé
def grad(f):
    """différenciation automatique"""
    g = autograd.grad
    def grad_f(x, y):
        return np.array([g(f, 0)(x, y), g(f, 1)(x, y)])
    return grad_f

def J(f):
    j = autograd.jacobian
    def J_f(x, y):
        return np.array([j(f, 0)(x, y), j(f, 1)(x, y)]).T
    return J_f

def Newton(F, x0, y0, eps=eps, N=N):
    x0, y0 = float(x0), float(y0)        
    for i in range(N):
        I= J(F)(x0,y0)
        (x, y) = (x0, y0) - np.dot(np.linalg.inv(I),F(x0, y0))
        if np.sqrt((x - x0)**2 + (y - y0)**2) <= eps:
            return [x, y]
        x0, y0 = x, y
    else:
        raise ValueError(f"no convergence in {N} steps.")
```

#### Question 6 + Tâche 3:
Comment, en partant d'un point de référence $(x_0, y_0)$ tel que $f(x_0, y_0)=c$, peut-on générer avec la méthode de Newton un point $(x_1, y_1)$ également tel que $f(x_1, y_1) = c$, mais à une distance $\delta > 0$ de $(x_0, y_0)$ et qui soit "à droite" quand on est en $(x_0, y_0)$ et qu'on regarde dans la direction de $\nabla f(x_0, y_0)$ ? Implémenter la fonction `level_curve` qui répète ce procédé $N-1$ fois et renvoie un tableau NumPy de taille `(2, N)` contenant les coordonnées des points correspondants, puis valider graphiquement le résultat au moyen des exemples de référence.

#### Réponse: 

On a un point de référence $(x_0, y_0)$ tel que $f(x_0, y_0)=c$ et on cherche un point $(x_1, y_1)$ tel que :
$$
\left\{
    \begin{array}{ll}
        f(x_1, y_1)=c \\
        (x_1, y_1)\: est \: à \: droite \: de \: \nabla f(x_0, y_0) \\
        d((x_1, y_1),(x_0, y_0))=\sqrt{(x_1-x_0)^{2}+(y_1-y_0)^{2}}=\delta
    \end{array}
\right.
$$
Pour faire cela, on va ajouter au point $(x_0, y_0)$ le vecteur $u$ de direction celle de $\nabla f(x_0, y_0)$ "tourné vers la droite", c'est-à-dire de $\pi/2$ et de norme $\delta$.
Notons ce nouveau point $(x', y')$.
On définit la fonction $F(x,y)= \begin{pmatrix} f(x,y) - c \\ \sqrt{(x_1-x_0)^{2}+(y_1-y_0)^{2}} - delta \end{pmatrix}$
parce qu'on veut 
\begin{array}{ll}
        f(x_1, y_1)=c \\
        d((x_1, y_1),(x_0, y_0))=\sqrt{(x_1-x_0)^{2}+(y_1-y_0)^{2}}=\delta
    \end{array}

On applique ensuite la fonction de Newton à $F$ en $(x', y')$ pour obtenir $(x_1, y_1)$

```python
def level_curve(f, x0, y0, delta=0.1, N=1000, eps=eps):
    c=f(x0,y0)
    grad_f = grad(f)
    xi=float(x0)                                    #on s'assure qu'on utilise bien des doubles car on va utiliser une fonction générée par autograd
    yi=float(y0)
    X,Y=[xi],[yi]
    for i in range(N):
        gradf=grad_f(xi,yi)
        u=np.dot(np.array([[0,-1],[1,0]]),gradf)
        u=u/np.linalg.norm(u)
        def F(x,y):
            return np.array([f(x,y) - c, np.sqrt((X[-1]-x)**2 + (Y[-1]-y)**2) - delta])
        xi+=u[0]*delta
        yi+=u[1]*delta
        xi,yi=Newton(F, xi, yi,eps,N)
        X+=[xi]
        Y+=[yi]
    M=np.array([X,Y])
    return(M)

###validation graphique

def veriftache3 (f,x0,y0,delta=0.1, N=1000, eps=eps): ###on représente graphiquement les points obtenus avec la fonction level_curve
    M=level_curve(f, x0, y0, delta=0.1, N=1000, eps=eps)
    plt.scatter(M[0],M[1],c="red")
```

#### Question 7 + Tâche 4

Proposer une nouvelle implémentation de `level_curve` qui arrête l'algorithme et renvoie les points connus quand le segment formé par les deux derniers points produits intersecte l'un des segments produits auparavant. En étant (raisonnablement) optimiste, doit-t'on prendre la peine de tester l'intersection du dernier segment avec tous les segments déjà produits (ce qui prend du temps ...) ou juste avec le premier ? Pour répondre à cette question, on pourra se demander si les courbes de niveau similaires à celle de la fonction de Rosenbrock passant par le point $(0.5, 0)$ (cf. Exemples de référence) -- c'est-à-dire "auto-intersectantes" -- sont fréquentes ou rares et pourquoi.

```python
def intersection(pointA,pointB,pointC,pointD): #les points sont des matrices colonne
    (xA,yA)=(pointA[0],pointA[1])
    (xB,yB)=(pointB[0],pointB[1])
    (xC,yC)=(pointC[0],pointC[1])
    (xD,yD)=(pointD[0],pointD[1])
    if xA>xB:
        (xA,yA,xB,yB)=(xB,yB,xA,yA)
    if xC>xD:
        (xC,yC,xD,yD)=(xD,xD,xC,yC)
    a=(yB-yA)/(xB-xA)                   #on calcule les coeffs directeurs
    b=(yD-yC)/(xD-xC)
    if a!=b:                            #les droites se croisent on regarde ou
        x=(yB-yD+b*xD-a*xB)/(b-a)
        if max(xA,xC)<=x<=min(xB,xD):   #il y a alors intersection
            return(True)
        else:
            return(False)
    if a==b:                             #les droites sont colinéaires
        if yB-a*xB!=yD-b*xD:             #on compare les ordonnées à l'origine - dans ce cas les droites ne sont pas confondues
            return(False)
        if yB-a*xB==yD-b*xD:             #les droites sont confondues
            if xB<xC:
                return(False)
            if xD<xA:
                return(False)
            else:
                return(True)

def level_curve2(f, x0, y0, delta=0.1, N=1000, eps=eps):
    c=f(x0,y0)
    grad_f = grad(f)
    xi=float(x0)                                    #on s'assure qu'on utilise bien des doubles car on va utiliser une fonction générée par autograd
    yi=float(y0)
    X,Y=[xi],[yi]
    for i in range(N):
        gradf=grad_f(xi,yi)
        u=np.dot(np.array([[0,-1],[1,0]]),gradf)
        u=u/np.linalg.norm(u)
        def F(x,y):
            return np.array([f(x,y) - c, np.sqrt((X[-1]-x)**2 + (Y[-1]-y)**2) - delta])
        xi+=u[0]*delta
        yi+=u[1]*delta
        xi,yi=Newton(F, xi, yi,eps,N)
        X+=[xi]
        Y+=[yi]
        if i>2 : 
             M=np.array([X,Y])
             pointA=M[:,0]
             pointB=M[:,1]
             pointC=M[:,i-1]
             pointD=M[:,i]
             if intersection(pointA,pointB,pointC,pointD):
                 return(M)
    M=np.array([X,Y])
    return(M)
```

```python
import autograd
import autograd.numpy as np

# Pandas
import pandas as pd

# Matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10] # [width, height] (inches). 

# Jupyter & IPython
from IPython.display import display

def display_contour(f, x, y, levels):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    fig, ax = plt.subplots()
    contour_set = plt.contour(
        X, Y, Z, colors="grey", linestyles="dashed", 
        levels=levels 
    )
    ax.clabel(contour_set)
    plt.grid(True)
    plt.xlabel("$x_1$") 
    plt.ylabel("$x_2$")
    plt.gca().set_aspect("equal")

def f1(x1, x2):
    x1 = np.array(x1)
    x2 = np.array(x2)
    return 3.0 * x1 * x1 - 2.0 * x1 * x2 + 3.0 * x2 * x2 

def f2(x1, x2):
    return (x1 - 1)**2 + (x1 - x2**2)**2

def f3(x, y):
    return np.sin(x + y) - np.cos(x * y) - 1 + 0.001 * (x * x + y * y) 


N=100
eps=10**(-3)


def grad(f):
    """différenciation automatique"""
    g = autograd.grad
    def grad_f(x, y):
        return np.array([g(f, 0)(x, y), g(f, 1)(x, y)])
    return grad_f

def J(f):
    j = autograd.jacobian
    def J_f(x, y):
        return np.array([j(f, 0)(x, y), j(f, 1)(x, y)]).T
    return J_f

def Newton(F, x0, y0, eps=eps, N=N):
    x0, y0 = float(x0), float(y0)        
    for i in range(N):
        I= J(F)(x0,y0)
        (x, y) = (x0, y0) - np.dot(np.linalg.inv(I),F(x0, y0))
        if np.sqrt((x - x0)**2 + (y - y0)**2) <= eps:
            return x, y
        x0, y0 = x, y
    else:
        raise ValueError(f"no convergence in {N} steps.")

def intersection(pointA,pointB,pointC,pointD): #les points sont des matrices colonne
    (xA,yA)=(pointA[0],pointA[1])
    (xB,yB)=(pointB[0],pointB[1])
    (xC,yC)=(pointC[0],pointC[1])
    (xD,yD)=(pointD[0],pointD[1])
    if xA>xB:
        (xA,yA,xB,yB)=(xB,yB,xA,yA)
    if xC>xD:
        (xC,yC,xD,yD)=(xD,xD,xC,yC)
    a=(yB-yA)/(xB-xA)                   #on calcule les coeffs directeurs
    b=(yD-yC)/(xD-xC)
    if a!=b:                            #les droites se croisent on regarde ou
        x=(yB-yD+b*xD-a*xB)/(b-a)
        if max(xA,xC)<=x<=min(xB,xD):   #il y a alors intersection
            return(True)
        else:
            return(False)
    if a==b:                             #les droites sont colinéaires
        if yB-a*xB!=yD-b*xD:             #on compare les ordonnées à l'origine - dans ce cas les droites ne sont pas confondues
            return(False)
        if yB-a*xB==yD-b*xD:             #les droites sont confondues
            if xB<xC:
                return(False)
            if xD<xA:
                return(False)
            else:
                return(True)

def level_curve2(f, x0, y0, delta=0.1, N=1000, eps=eps):
    c=f(x0,y0)
    grad_f = grad(f)
    xi=float(x0)                                    #on s'assure qu'on utilise bien des doubles car on va utiliser une fonction générée par autograd
    yi=float(y0)
    X,Y=[xi],[yi]
    for i in range(N):
        gradf=grad_f(xi,yi)
        u=np.dot(np.array([[0,-1],[1,0]]),gradf)
        u=u/np.linalg.norm(u)
        def F(x,y):
            return np.array([f(x,y) - c, np.sqrt((X[-1]-x)**2 + (Y[-1]-y)**2) - delta])
        xi+=u[0]*delta
        yi+=u[1]*delta
        xi,yi=Newton(F, xi, yi,eps,N)
        X+=[xi]
        Y+=[yi]
        if i>2 : 
             M=np.array([X,Y])
             pointA=M[:,0]
             pointB=M[:,1]
             pointC=M[:,i-1]
             pointD=M[:,i]
             if intersection(pointA,pointB,pointC,pointD):
                 return(M)
    M=np.array([X,Y])
    return(M)



###validation graphique

def tache5 (f,x0,y0,delta=0.1, N=1000, eps=eps): ###on représente graphiquement les points obtenus avec la fonction level_curve
    M=level_curve2(f, x0, y0, delta=0.1, N=1000, eps=eps)
    plt.scatter(M[0],M[1],c="red")
```

## Post-traitement

Pour améliorer le rendu des courbes de niveau, il est tentant de diminuer la distance $\delta$ entre deux points consécutifs. Malheureusement, dans des cas d'usage réalistes, il est fréquent que la fonction $f$ soit beaucoup plus complexe que les exemples qui nous avons considérés. Le nombre de fois où cette fonction est appelée est alors le principal facteur qui détermine le temps de calcul des lignes de niveau. On cherche donc dans cette section à améliorer la qualité visuelle des résultats sans faire exploser le nombre d'évaluation de la fonction $f$. On va chercher à résoudre ce problème en reliant les points successifs par des chemins plus "flexibles" que des segments de droites, ayant le potentiel de mieux coller aux lignes de niveau.

### Interpolation 

#### Question 8:

Soit $P_1$ et $P_2$ deux points du plan et $u_1$ et $u_2$ deux vecteurs du plan non nuls.
On cherche à produire un chemin $\gamma: [0,1] \to \mathbb{R}^2$ continûment différentiable, joignant $P_1$ et $P_2$ ($\gamma(0) = P_1$ et $\gamma(1) = P_2$), tel que $\gamma'(0)$ soit dans la même direction et le même sens que $u_1$ et $\gamma'(1)$ soit dans la même direction et le même sens que $u_2$.

On recherche une telle solution sous la forme
$\gamma(t) = (x(t), y(t))$ où 
$$
x(t) = a + bt + ct^2 \; \mbox{ et } \; y(t) = d + et + ft^2
$$
et les paramètres réels $a, b, c, d, e, f$ sont à déterminer(*).


Déterminer les coefficients $a, b, c, d, e, f$ solutions de ce problème quand ils existent.
Expliciter si nécessaire les conditions que doivent remplir $P_1$, $P_2$, $u_1$ et $u_2$ pour qu'une solution existe.

(*) Cf. ["Mathematical Typography" de Donald E. Knuth](https://projecteuclid.org/euclid.bams/1183544082) pp. 354-359 (Bull. Amer. Math. Soc. (N.S.), Volume 1, Number 2 (1979), 337-372), pour une classe de chemins interpolants similaire mais plus complexe.

#### Réponse: 

On suppose qu'une telle solution existe. On pose $P_1 = (p_{x1},p_{y1}), \: P_2 = (p_{x2},p_{y2}), \: u_1 = (u_{x1},u_{y1})$ et $u_2 = (u_{x2},u_{y2})$. Alors, on a :

- $\gamma(0) = P_1$ donc:
$$
\begin{pmatrix}
   p_{x1}  \\
   p_{y1}
\end{pmatrix}=\begin{pmatrix}
   a \\
   d
\end{pmatrix}
$$

- De même, $\gamma(1) = P_2$ donc:
$$
\begin{pmatrix}
   p_{x2}  \\
   p_{y2}
\end{pmatrix}=\begin{pmatrix}
   a+b+c \\
   d+e+f
\end{pmatrix}
$$

On s'intéresse à présent aux dérivées, et comme on veut que $\gamma'(0)$ et $u1$ soient colinéaires on introduit $\lambda \in \mathbb{R}$ tel que $\gamma'(0)=\lambda.u1$. De même, on introduit $\mu \in \mathbb{R}$ tel que $\gamma'(1)= \mu.u2$. Ainsi, on a:

- $b = \lambda*u_{x1}$ et $e = \lambda*u_{y1}$
- $b + 2c = \mu*u_{x2}$ et $e + 2f = \mu*u_{y2}$

On obtient donc le système à 8 équations suivant:
$$
\left\{
    \begin{array}{ll}
         p_{x1}=a \\
         p_{y1}=d \\
         p_{x2}= a+b+c \\
         p_{y2}= d+e+f \\
         b=\lambda \times u_{x1} \\
         e = \lambda \times u_{y1} \\
         b + 2c = \mu \times u_{x2} \\
         e + 2f = \mu \times u_{y2}
      \end{array}
\right.
$$
On résoud manuellement un tel système et on obtient alors que:
$$
\left\{
    \begin{array}{ll}
         p_{x1}=a \\
         p_{y1}=d \\
         \lambda = \frac{\frac{2*(x_2-x_1)}{u_{x2}} - \frac{2*(y_2-y_1)}{u_{y2}}}{\frac{u_{x1}}{u_{x2}} - \frac{u_{y1}}{u_{y2}}} \\
         b = \frac{\frac{2*(x_2-x_1)}{u_{x2}} - \frac{2*(y_2-y_1)}{u_{y2}}}{\frac{u_{x1}}{u_{x2}} - \frac{u_{y1}}{u_{y2}}} \times u_{x1} \\
         e = \frac{\frac{2*(x_2-x_1)}{u_{x2}} - \frac{2*(y_2-y_1)}{u_{y2}}}{\frac{u_{x1}}{u_{x2}} - \frac{u_{y1}}{u_{y2}}} \times u_{y1} \\
         \mu = \frac{2*(y_2-y_1) -\lambda*u_{y1}}{u_{y2}}
      \end{array}
\right.
$$

Et ainsi de suite...

Cependant, un tel système n'admet une solution que dans certaines conditions, que l'on trouve en faisant les calculs manuellement:
- $u_1$ et $u_2$ ne doivent pas être colinéaires pour ne pas annuler le dénominateur de $\lambda$
- Pour que le numérateur de $\lambda$ soit bien défini, il faut que $u_{x2}$ et $u_{y2}$ soient tous deux non nuls.
- Pour que $\gamma'(0)$ et $u_1$ soient dans le même sens, on doit avoir $\lambda > 0$, et donc $\frac{2*(x_2-x_1)}{u_{20}} -{2*(y_2-y_1)}{u_{21}}$ et $\frac{u_{10}}{u_{20}} - \frac{u_{11}}{u_{21}}$ de même signe.
- Pour que $\gamma'(1)$ et $u_2$ soient dans le même sens, on doit avoir $\mu > 0$, et donc $2*(y_2-y_1) -\lambda*u_{y1}$ et $u_{y2}$ du même signe. 

#### Tâche 6

Implémenter la solution sous la forme d'une fonction `gamma` dont les arguments sont `t`, `P1`, `P2`, `u1` et `u2` et qui renvoie le ou les points $\gamma(t)$ associés. Lorsqu'il n'existe pas de chemin de la forme souhaitée pour les paramètres `P1`, `P2`, `u1` et `u2`, on utilisera comment remplacement de $\gamma$ un chemin rectiligne interpolant linéairement les points $P_1$ et $P_2$.

Pour des raisons de performance, on vectorisera cette fonction par rapport à `t` : 

  - en acceptant comme argument `t` des tableaux NumPy (monodimensionels) de nombres flottants et en renvoyant alors un tableau de taille `(2, len(t))` flottants décrivant l'abscisse et l'ordonnée des `len(t)` points $\gamma(t)$ correspondant,

  - en appliquant directement opérateurs et fonctions mathématiques aux tableaux NumPy, sans utiliser de boucle `for`,
  
On validera ensuite graphiquement l'implémentation sur un exemple où l'on représentera les point $P_1$ et $P_2$, les tangentes associées et le chemin $\gamma$ correspondant.




```python
def gamma(T,P1,P2,u1,u2):
    """une fonction vectorisée où T est un array"""
    x1,y1,x2,y2 = P1[0],P1[1],P2[0],P2[1]
    ux1,uy1,ux2,uy2= u1[0],u1[1],u2[0],u2[1]
    droite = list((np.add(x1,np.multiply(T,(x2-x1))),np.add(y1,np.multiply(T,(y2-y1)))))
    if ux2==0 or uy2==0 or np.cross(u1,u2)==0: #si aucune solution n'existe
        return np.array(droite)
    else:
        l_x = (2*(x2-x1)/ux2 - 2*(y2-y1)/uy2)/(ux1/ux2 - uy1/uy2)
        m_x = (2*(x2-x1)-l_x*ux1)/ux2
        if l_x <= 0 or m_x <= 0:
            return np.array(droite) #gamma'(0) et u1 ou gamma'(1) et u2 n'ont pas le même sens 
        else:
            a = x1 #soltions du système calculées précédemment
            b = l_x*ux1
            c = x2 - b - a
            d = y1
            e = l_x*uy1
            f = y2 - e - d
            x = np.add(np.add(a, np.multiply(T,b)),np.multiply(c,np.power(T,2)))
            y = np.add(np.add(d, np.multiply(T,e)),np.multiply(f,np.power(T,2)))
            vect = list((x,y))
            return np.array(vect)

###Validation graphique
P1=(1.0,1.0)
P2=(1.25,1.75)
plt.plot([1.0],[1.0],'ro', markersize='8') #on dessine les points sur le graphique
plt.plot([1.25],[1.75],'ro', markersize='8')
u1=(-5,4)
u2=(2,1)
X = np.linspace(0,1.5,100)
Y= gamma(X,P1,P2,u1,u2)
plt.plot(Y[0],Y[1])

def tangente1(x):
    return (u1[1]/u1[0])*(x-P1[0])+P1[1]

def tangente2(x):
    return (u2[1]/u2[0])*(x-P2[0])+P2[1]

Y1=[tangente1(x) for x in X]
Y2=[tangente2(x) for x in X]
plt.plot(X,Y2,"r--")
plt.plot(X,Y1,"r--")
plt.title("chemin entre deux points")
plt.show()
```
